```python
def is_palindrome(num):
    temp = num
    rev = 0

    while temp > 0:
        digit = temp % 10
        rev = rev * 10 + digit
        temp //= 10

    return rev == num
```

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result
```

```python
def is_armstrong(n):
    digits = str(n)
    power = len(digits)

    total = 0
    for d in digits:
        total += int(d) ** power

    return total == n
```

```python
import math

n = input("Enter a number: ")

if not n.lstrip("-").isdigit():
    print("Invalid input")
else:
    n = int(n)

    if n <= 1:
        print("Neither prime nor composite")
    else:
        is_prime = True
        for i in range(2, int(math.sqrt(n)) + 1):
            if n % i == 0:
                is_prime = False
                break

        if is_prime:
            print("Prime")
        else:
            print("Composite")
```

```python
def is_perfect(n):
    total = 0

    for i in range(1, n):
        if n % i == 0:
            total += i

    return total == n
```
