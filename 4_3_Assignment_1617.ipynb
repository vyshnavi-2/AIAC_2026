```python
def is_leap_year(year):
    if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
        return True
    else:
        return False
```

```python
print(is_leap_year(2024))
print(is_leap_year(2023))
```

```python
def cm_to_inches(cm):
    return round(cm / 2.54, 2)
```

```python
print(cm_to_inches(10))
print(cm_to_inches(25))
```

```python
def format_name(full_name):
    parts = full_name.split()
    first = parts[0]
    last = parts[-1]
    return f"{last}, {first}"
```

```python
print(format_name("John Smith"))
print(format_name("Anita Rao"))
print(format_name("Rahul Verma"))
```

```python
def count_vowels_zero(text):
    count = 0
    for ch in text.lower():
        if ch in "aeiou":
            count += 1
    return count
```

```python
def count_vowels_few(text):
    vowels = "aeiouAEIOU"
    return sum(1 for ch in text if ch in vowels)
```

```python
print(count_vowels_zero("Education"))
print(count_vowels_few("Education"))
```

```python
def count_lines(filename):
    with open(filename, "r") as file:
        return len(file.readlines())
```

```python
# Create a dummy sample.txt file for demonstration
with open("sample.txt", "w") as f:
    f.write("line 1\n")
    f.write("line 2\n")
    f.write("line 3\n")

print(count_lines("sample.txt"))
```
